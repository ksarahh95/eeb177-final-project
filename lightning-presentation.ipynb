### Introduction

For this final project, I looked specifically at the genus Canis of the family Canidae, that consists of various extant species such as dogs, coyotes, and wolves. Species of this genus typically possess long legs, well-developed skulls, moderate to large body size, and short tails and ears. The primary focus of the project was to study which species was the most prominent in the Canis genus and the diversity of the group over time, using different tools of analysis. From literature, it was known that there was an interesting explosion of Canis evolution during the Early Pleistocene period across Eurasia, called the Wolf event. The explosion was most likely the result of continental glaciation and the Mammoth steppe formation. In another study, it was discovered that Canis diversity decreased towards the latter half of the Early Pleistocene to the Middle Pleistocene periods.

### 1. Find the number of unique canis species.

```python
#output name of species along with the rank "species."
def species_list(filename): #function species_list
    dog_species = {} 
    ff = open(filename, "r", encoding = "ISO-8859-15")
    ss = ff.readlines()[3:] #read file from line 2
    for line in ss:
        line = line.split(",")
        name = line[5] #read 5th column
        rank = line[6] #read 6th column
        if rank == 'species':
            dog_species[name] = rank #matching name to rank
    return dog_species
species_list("my-final-formatted-dog.csv")

#87 unique species
```

### 3. Make a list of time periods (early interval) where Canis species were found.

```python
#output list of species names alongside the early interval time periods and occurrences
def list_of_periods(filename): #create function list_of_periods
    TimePeriods = []
    periods_dog = open(filename, "r", encoding = "ISO-8859-15") #read file
    list_dog = periods_dog.readlines()[3:]
    for line in list_dog:
        name = line.split(",")[5]      #lists species names
        early = line.split(",")[12]    #lists early time periods
        occurrences = line.split(",")[0] #lists occurrences of species
        print(str(name) + ", " + str(early))
        print(occurrences)
    return TimePeriods
list_of_periods("my-final-formatted-dog.csv") 
```

### 4. Find number of species over time.

```python
def Canis():
    CanisCSV= open("/home/eeb177-student/Desktop/eeb-177/eeb177-final-project/my-final-formatted-dog.csv", "r", encoding = "ISO-8859-1")
    alldata=CanisCSV.readlines() [3:] #start reading from line 3
    Time_dict = {} #create an empty dictionary
    for line in alldata:
        record_elements = line.split(",") 
        if len(record_elements) < 12:
            continue
        Timeperiod = record_elements[12] #read column 12
        if Timeperiod in Time_dict: 
            count = Time_dict[Timeperiod] #extract count value for the time period
            Time_dict[Timeperiod] = count + 1 #add one to the count 
        else: 
            Time_dict[Timeperiod] = 1 #when time period is new assign value of 1           
    return Time_dict
```

```python
print (Canis())
```

```python
import matplotlib.pyplot as plot
plot.figure(figsize=(22,15))
plot.bar(range(len(Canis())), Canis().values(), align='center', color = 'purple')
plot.xticks(range(len(Canis())), list(Canis().keys()), rotation = 'vertical')
plot.margins(0.01)
plot.subplots_adjust(bottom=0.1)
plot.title('Number of Species Across Time', fontsize = 35)
plot.xlabel('Time Period', fontsize = 35)
plot.ylabel('Number of Species', fontsize = 35)
plot.show()
```

### 5. Make a list of species along with their first and last occurrences.

```python
def Min_ma(filename):
    minage_dict = {} 
    species_data = open(filename, "r", encoding = "ISO-8859-15") 
    minage_records = species_data.readlines()[3:] 
    for line in minage_records:
        line = line.replace('\"', '') 
        minage_elements = line.split(",")
        species = minage_elements[5] #read column 5
        starting_period = float(minage_elements[14]) #read column 14
        minage_dict[species] = starting_period
    return minage_dict

fo_dict = Min_ma("my-final-formatted-dog.csv") #list species names with first occurrence
fo_dict
```

```python
def Max_ma(filename):
    maxage_dict = {} 
    rspecies_data = open(filename, "r", encoding = "ISO-8859-15") 
    maxage_records = rspecies_data.readlines()[3:] 
    for line in maxage_records:
        line = line.replace('\"', '') 
        maxage_elements = line.split(",")
        species = maxage_elements[5] #read column 5
        ending_period = float(maxage_elements[15]) #read column 15
        maxage_dict[species] = ending_period 
    return maxage_dict

lo_dict = Max_ma("my-final-formatted-dog.csv")        #list species along with last occurrence
lo_dict
```
