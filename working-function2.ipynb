# Function 2: Making a list of time periods

```python
def list_of_periods(filename):
    TimePeriods = []
    periods_dog = open(filename, "r", encoding = "ISO-8859-15")
    list_dog = periods_dog.readlines()[2:]
    for line in list_dog:
        name = line.split(",")[5]      #lists species names
        early = line.split(",")[12]    #lists early time periods
        late = line.split(",")[13]     #lists late time periods
        print(str(name) + ", " + str(early) + ", " + str(late))
    return TimePeriods

```

```python
list_of_periods("my-final-formatted-dog.csv")
```
